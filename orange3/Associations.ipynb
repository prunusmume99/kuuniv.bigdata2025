```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(df1, min_support=.01, max_len=6, use_colnames=True)

frequent_itemsets.sort_values(by=['support']).head()

```

```python
rules = association_rules(frequent_itemsets, metric='support', min_threshold=0.01)

c1 = rules['confidence'] > 0.85
filtered_rules = rules[c1]

filtered_rules.sort_values(by=['support'], ascending=False)
```
